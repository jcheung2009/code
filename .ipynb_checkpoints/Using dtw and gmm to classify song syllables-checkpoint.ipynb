```matlab
addpath('~/code/evsonganaly/')
addpath('~/code/from_evren/')

%get sample song data from a bird's song
pathnm = '/opt/data/deaf_and_naspm/rd36gr25/saline_2_11_2018/';
batchlist = 'batch.keep';

%iterate through first 20 song files in directory and extract onsets, offsets, labels, and raw amplitude 
%waveform for each song
ff = load_batchf([pathnm,batchlist]);
ons = []; offs = []; lbls = []; filtsong = [];
tm = 0; 
for i = 1:20
    load([pathnm,ff(i).name,'.not.mat']);
    ons = [ons;onsets+tm];
    offs = [offs;offsets+tm];
    lbls = [lbls labels];
    [dat fs] = evsoundin(pathnm,ff(i).name,'obs0');
    filtsong = [filtsong; bandpass(dat,fs,300,10000,'hanningffir')];
    tm = tm+(length(dat)*1000)/fs;
end
```

```matlab
%spectrogram parameters
fs = 32000;%sampling rate for audio files 
buff = 0.016*fs;%16 ms buffer at onsets and offsets
NFFT = 512;
overlap = NFFT-32;%1 ms 
t=-NFFT/2+1:NFFT/2;
sigma=(1/1000)*fs;
w=exp(-(t/sigma).^2);%gaussian smoothing window 

%set of syllables that song was manuall labelled 
sylls = unique(lbls);
sylls = sylls(isletter(sylls))
```

```matlab
%extract spectrograms of 10 renditions of each song syllable and visualize an example spectrogram for each syllable
templates = {};
figure;hold on;
for i = 1:length(sylls)
    ind = strfind(lbls,sylls(i));
    ix = ind(randsample(length(ind),10));
    for m = 1:length(ix)
        dat = filtsong(floor(ons(ix(m))*fs/1000)-buff:ceil(offs(ix(m))*fs/1000)+buff);
        sp = spectrogram(dat,w,overlap,NFFT,fs);
        sp = abs(sp);
        if m == 1
            subplot(1,length(sylls),i);hold on;
            sp2 = sp;
            ind = find(sp2<=mean(mean(sp2)));
            sp2(ind) = mean(mean(sp2))/2;
            imagesc(log(sp2));axis('xy');title(num2str(sylls(i)));set(gca,'XTickLabel',[],'YTickLabel',[])
        end
        sp = sp./sum(sp,2);%normalize the spectrogram at each time slice
        templates = [templates; sp];
    end
end
```

```matlab
%calculate distrance metric from dynamic time warp to each of the template spectrograms 
%for every syllable instance
dist = NaN(length(lbls),length(templates));
for i = 1:length(templates)
    for ii = 1:length(lbls)
        dat = filtsong(floor(ons(ii)*fs/1000)-buff:ceil(offs(ii)*fs/1000)+buff);
        sp = spectrogram(dat,w,overlap,NFFT,fs);
        sp = abs(sp);
        sp = sp./sum(sp,2);%normalize the spectrogram at each time slice
        dist(ii,i) = dtw(templates{i},sp);
    end
end
```

```matlab
%feature matrix is mxn where m is observation and n is number of features = 80
disp([num2str(size(dist,1)),' observations each with ',num2str(size(dist,2)),' features']);
dist2 = dist;

%pca to reduce dimensionality of feature vectors
[c,s,~,~,explained] = pca(dist2);
ind_comp = find(cumsum(explained)<99);
dist2 = dist2*c(:,ind_comp);%project matrix on principal component space
disp([num2str(length(ind_comp)),' principal components']);
```

```matlab
%view example of syllable clusters in pca space 
cmbs = combnk([1:length(ind_comp)],2);%all combinations of components
figure;hold on;
for i = 1:16;
    subplot(4,4,i);hold on;
    plot(dist2(:,cmbs(i,1)),dist2(:,cmbs(i,2)),'k.');
    xlabel(['comp ',num2str(cmbs(i,1))]);
    ylabel(['comp ',num2str(cmbs(i,2))]);
end
```

```matlab
%train gmm on a range of models with different components and compare BIC/AIC to identify in an unsupervised 
%fashion the ideal model/number of components

%divide training set into 3 subsets
k = 3;%number of sets
nrow = size(dist2,1);
n = floor(nrow/k);%three sets
lens = floor(nrow/n);
subset = cell(3,1);
for i = 1:k
    subset{i} = dist2((i-1)*n+1:i*n,:);
end

%train on the k data subsets
models = 2:16;%range of component numbers to test
options = statset('MaxIter',10000);
aic=[];bic=[];
for m = 1:length(models)
    bics = [];aics = [];
    for y = 1:k
        trainset = cell2mat(subset(y));
        gmm = fitgmdist(trainset,models(m),'Options',options,'Replicates',5,'SharedCovariance',true,...
        'CovarianceType','full','RegularizationValue',0.01);
        bics = [bics;gmm.BIC];
        aics = [aics; gmm.AIC];
    end
    aic=[aic;mean(aics)];
    bic=[bic;mean(bics)];
end

%plot AIC/BIC 
figure;subplot(1,2,1);hold on;plot(models,aic,'k.');ylabel('aic');xlabel('number of components');
subplot(1,2,2);hold on;plot(models,bic,'k.');ylabel('bic');xlabel('number of components');
[~,ind] = min(bic);
numcomponents = models(ind)+2;
disp([num2str(numcomponents),' components'])
```

```matlab
%train up gmm on chosen number of components
options = statset('MaxIter',100000);
gmm = fitgmdist(trainset,numcomponents,'Options',options,'Replicates',10,'SharedCovariance',true,...
'CovarianceType','full','RegularizationValue',0.01);
idx = cluster(gmm,dist2);%assign cluster number for each observation
```

```matlab
%visualize some of the spectrograms of the syllables within each cluster 
for i = 1:numcomponents
    if ~isempty((find(idx==i)))
        figure;hold on;
        onset=ons(find(idx==i));
        offset = offs(find(idx==i));
        if length(find(idx==i))>36
            numspecs = 36;
        else
            numspecs = length(find(idx==i));
        end
        for ix = 1:numspecs 
            subplot(6,6,ix);hold on;
            dat = filtsong(floor(onset(ix)*fs/1000)-buff:ceil(offset(ix)*fs/1000)+buff);
            [sp f] = spectrogram(dat,w,overlap,NFFT,fs);
            ind = find(f>=300 & f<=10000);%restrict spectrogram to within bandpass frequencies
            sp = abs(sp(ind,:));
            ind = find(abs(sp)<=mean(mean(abs(sp))));
            sp(ind)=mean(mean(abs(sp)))/2;
            imagesc(log(sp));axis('xy');title(num2str(i));set(gca,'XTickLabel',[],'YTickLabel',[])
        end
    end
end
```

```matlab
for i = 1:numcomponents
    disp(['labels in cluster ',num2str(i),':',lbls(find(idx==i))])
end
```

```matlab
%store mapping of syllable label and cluster number in structure
lbl_to_cluster = struct('a',2,'b',[4,8,9],'c',[5,7,11,12],'d',16,'e',13,'f',1,'s',6,'w',[3,10,14,17,18]);
```

```matlab
%validate trained gmm on new set of syllables and compare with manual labels to get % correct 

%extract syllables from another set of 20 songs that were not trained by gmm 
ons2 = []; offs2 = []; lbls2 = []; filtsong2 = [];
tm2 = 0; 
for i = 31:50 
    load([pathnm,ff(i).name,'.not.mat']);
    ons2 = [ons2; onsets+tm2];
    offs2 = [offs2; offsets+tm2];
    lbls2 = [lbls2 labels];
    [dat fs] = evsoundin(pathnm,ff(i).name,'obs0');
    filtsong2 = [filtsong2; bandpass(dat,fs,300,10000,'hanningffir')];
    tm2= tm2+(length(dat)*1000)/fs;
end

%compute the distance metric to the 80 template syllables selected previously 
dist_val = NaN(length(lbls2),length(templates));
for i = 1:length(templates)
    for ii = 1:length(lbls2)
        dat = filtsong2(floor(ons2(ii)*fs/1000)-buff:ceil(offs2(ii)*fs/1000)+buff);
        sp = spectrogram(dat,w,overlap,NFFT,fs);
        sp = abs(sp);
        sp = sp./sum(sp,2);
        dist_val(ii,i) = dtw(templates{i},sp);
    end
end
```

```matlab
disp([num2str(size(dist_val,1)),' observations in validation set'])
```

```matlab
%project distance matrix on pca space defined previously
dist2_val = dist_val;
dist2_val = dist2_val*c(:,ind_comp);

%view example of syllable clusters in pca space 
figure;hold on;
for i = 1:16;
    subplot(4,4,i);hold on;
    plot(dist2_val(:,cmbs(i,1)),dist2_val(:,cmbs(i,2)),'k.');
    xlabel(['comp ',num2str(cmbs(i,1))]);
    ylabel(['comp ',num2str(cmbs(i,2))]);
end
```

```matlab
%classify each syllable based on gmm clustering
idx = cluster(gmm,dist2_val);

%visualize the spectrograms in each cluster
for i = 1:numcomponents
    if ~isempty((find(idx==i)))
        figure;hold on;
        onset=ons2(find(idx==i));
        offset = offs2(find(idx==i));
        if length(find(idx==i))>36
            numspecs = 36;
        else
            numspecs = length(find(idx==i));
        end
        for ix = 1:numspecs 
            subplot(6,6,ix);hold on;
            dat = filtsong2(floor(onset(ix)*fs/1000)-buff:ceil(offset(ix)*fs/1000)+buff);
            [sp f] = spectrogram(dat,w,overlap,NFFT,fs);
            ind = find(f>=300 & f<=10000);%restrict spectrogram to within bandpass frequencies
            sp = abs(sp(ind,:));
            ind = find(abs(sp)<=mean(mean(abs(sp))));
            sp(ind)=mean(mean(abs(sp)))/2;
            imagesc(log(sp));axis('xy');title(num2str(i));set(gca,'XTickLabel',[],'YTickLabel',[])
        end
    end
end
```

```matlab
%examine the manual labels within each cluster of validation set
for i = 1:numcomponents
    disp(['labels in cluster ',num2str(i),':',lbls2(find(idx==i))])
end
```

```matlab
%for each syllable, compute percentage that was correctly classified based on manual labels
perc_correct = [];
for i = 1:length(sylls)
    ids = lbl_to_cluster.(sylls(i));
    clustered_lbls = [];
    for m = 1:length(ids)
        clustered_lbls = [clustered_lbls lbls2(find(idx==ids(m)))];
    end
    perc_correct = [perc_correct;length(strfind(clustered_lbls,sylls(i)))/length(clustered_lbls)];
end
perc_correct
disp(['average % correct: ',num2str(mean(perc_correct))])
```
